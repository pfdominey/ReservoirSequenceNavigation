```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import pingouin as pg
from numpy.linalg import norm
from itertools import combinations

from plotter import *
from plotnine import *
from easyesn.optimizers import GradientOptimizer
from easyesn import PredictionESN
from easyesn import BaseESN
# note: you must make this change in the easyesn code
# https://github.com/kalekiu/easyesn/issues/12
from easyesn.optimizers import GridSearchOptimizer
from easyesn import helper as hlp
import ipywidgets as widgets
from IPython.display import display, Markdown, Latex
import matplotlib
import time
import functools
import copy

# CREATE DUMMY DATA
sample_xs = np.arange(50) * 0.1
sample_ys = np.tile( np.concatenate([np.sin(sample_xs), np.cos(sample_xs)]), 20) 
sample_noisy_ys = sample_ys + np.random.uniform(-0.3, 0.3, sample_ys.shape)
sample_xs = np.tile( sample_xs,  40)
sample_groups = np.tile(['noisy sin'] * 50 + ['noisy cos']*50, 20)


df = pd.DataFrame({'group':sample_groups, 'xs': sample_xs, 'ys':sample_noisy_ys})

display(Markdown("### Sample dataframe"))
display(df.sample(8))

fig = plt.figure()
plt.title('standard deviations')
sns.lineplot(x="xs", y="ys",  data = df, ci='sd', hue='group')
display(fig)
plt.close(fig)


fig = plt.figure()
sns.lineplot(x="xs", y="ys",  data = df, hue='group')
plt.title('Confidence intervals')
display(fig)
plt.close(fig)
```
